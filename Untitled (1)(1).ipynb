```python
# Function that gets a row from rdd and return a splited array
# of lowerCased words in it.

def SplitToLower(lines):
      return lines.lower().split()
```

```python
# Function that gets a word and deletes all the non-letter symbols.

def OmitNonWords(word):
    newWord = ''
    
    for i in range(len(word)):
        if (word[i] >= 'a' and word[i] <= 'z') or word[i] == "'":
            newWord += word[i]
    
    if (newWord != '' and newWord[0] == newWord[-1] and newWord[0] == "'"):
        newWord = newWord[1:-2]
    if (newWord != '' and newWord.count("'") >= 1):
        newWord = newWord[0:newWord.index("'")]
    if (newWord != '' and newWord.count(newWord[0]) == len(newWord)):
        newWord = ''
        
    return newWord
```

```python
# Function that checks if the given word is a 'stop word'

def IsNotStopWord(word):
    if word in stops:
        return False
    
    return True
```

```python
# Creating an rdd (that looks like a dictionary) from all the files
# Every row in the rdd is a tuple of a word and a number of the file
# that this word was found in it.

dic = sc.parallelize([])

for i in range(len(bookNames)):
    bookPath = os.path.join("./books", bookNames[i])
    bookRDD = sc.textFile(bookPath)
    dic = dic.union(bookRDD.map(lambda x: x.replace("-", " ")) \
                           .flatMap(SplitToLower) \
                           .map(OmitNonWords) \
                           .filter(lambda x: len(x) > 1) \
                           .filter(IsNotStopWord) \
                           .map(lambda x: (x, i)))
```

```python
# Funtion that gets an iterable object, converts it to 
# a dictionary and returns it
def tfInnerFunc(x):
    arr = list(x)
    dic = {}
    
    for i in range(len(arr)):
        if (arr[i] in dic.keys()):
            dic[arr[i]] = dic[arr[i]] + 1
        else:
            dic[arr[i]] = 1
            
    return dic
```

```python
tfTable = dic.groupByKey().mapValues(tfInnerFunc)
```

```python
# Creating idf table
```

```python
idfTable = tfTable.map(lambda x: \
                       (x[0], \
                        x[1], \
                        math.log((len(bookNames)/len(x[1])),10)))
```

```python
idfTable.take(10)
```

```python
# Creating tf_idf table
```

```python
# Function that gets a row '(word, dictionary, idf_value)',
# calculates the tf_idf_values for every word <-> file cuple
# and returns a tuple '(word, dictionary_of_tf_idf_values)'

def tf_idfInnerFunc(x):
    dic = x[1]
    
    for key in dic:
        dic[key] = (1 + math.log(dic[key], 10)) * x[2]
        
    return (x[0], dic)
```

```python
tf_idfTable = idfTable.map(tf_idfInnerFunc)
```

```python
tf_idfTable.take(10)
```

```python
# Finding Cos-similarity, 5 nearest files to every file 
# and 5 nearest cuples of files
```

```python
# Function that gets a row as ('word', dictionary) and returns a row as
# tuples (fileNumber, 'word', tf-idf)

def aggregateInnerTable(x):
    dic = x[1]
    retArr = []
    
    for i in range(len(bookNames)):
        if (i in dic.keys()):
            retArr.append((i, x[0], dic[i]))
        else:
            retArr.append((i, x[0], 0))
            
    return retArr
```

```python
# Function that gets an itterable group of ('word', tf-idf)
# and returns a list with normalized tf-idf

def normalizeCos(x):
    arr = list(x)
    sqrSumDic = 0
    
    for i in range(len(arr)):
        sqrSumDic = sqrSumDic + (arr[i][2] * arr[i][2])

    sqrtSum = math.sqrt(sqrSumDic)
    
    if (sqrtSum != 0):
        for i in range(len(arr)):
            arr[i] = (arr[i][0], arr[i][1], arr[i][2] / sqrtSum)
    
    return arr
```

```python
# Function that convert an itterable group to a dictionary

def makeDict(x):
    arr = list(x)
    dic = {}
    
    for i in range(len(arr)):
        dic[arr[i][0]] = arr[i][2]
        
    return dic
```

```python
# Normalizing tf-idf

normalizedCosTable = tf_idfTable.map(aggregateInnerTable) \
    .flatMap(lambda x: x) \
    .groupBy(lambda x: x[0]) \
    .mapValues(normalizeCos) \
    .map(lambda x: x[1]) \
    .flatMap(lambda x: x) \
    .groupBy(lambda x: x[1]) \
    .mapValues(makeDict)
```

```python
# Function that gets a row ('word', dict) and deletes
# all the 0 tf-idf values from the dict

def deleteNulls(x):
    dic = x[1]
    newDic = {}
    
    for key in dic:
        if dic[key] != 0:
            newDic[key] = dic[key]
            
    return (x[0], newDic)
```

```python
# Deleting 0 tf-idf values and empty words

normalizedCosTable = normalizedCosTable.map(deleteNulls) \
    .filter(lambda x: len(x[1]) != 0)
```

```python
# Function that gets a row '(word, dictionary)'
# and returns a dictionary converted to a list

def tempConvertToTable(x):
    dic = x[1]
    arr = numpy.zeros((len(bookNames) + 1))
    
    for key in dic:
        arr[key] = dic[key]
        
    return arr
```

```python
# Function that get a list of normalized_tf_idf_values
# and index of file. Multiplies every values in a list
# by values of the given index file' value.

def tempMult(x, indx):
    for i in range(len(x)):
        x[i] = x[i] * x[indx]
        
    return x
```

```python
# Function that gets a row as ('word', dictionary) and returns a row as
# tuples (fileNumber, 'word', tf-idf)

def tempAggregateInnerTable(x):
    dic = x[1]
    retArr = []
    
    for i in range(len(bookNames) + 1):
        if (i in dic.keys()):
            retArr.append((i, x[0], dic[i]))
        else:
            retArr.append((i, x[0], 0))
            
    return retArr
```

```python
# Function that gets a string and returns a normalized
# table of the string and the other used files.

def createNormCosTable(string):
    tempStrArr = string.lower().split()
    strArr = []
    for i in range(len(tempStrArr)):
        tempStrArr[i] = OmitNonWords(tempStrArr[i])
        if (IsNotStopWord(tempStrArr[i]) and len(tempStrArr[i]) > 1):
            strArr.append(tempStrArr[i])
    
    for i in range(len(strArr)):
        strArr[i] = (strArr[i], len(bookNames))
    
    tempDic = dic.union(sc.parallelize(strArr))
    
    tempTfTable = tempDic.groupByKey().mapValues(tfInnerFunc)
    tempIdfTable = tempTfTable.map(lambda x: \
                       (x[0], \
                        x[1], \
                        math.log((len(bookNames)+1)/len(x[1]),10)))
    tempTf_idfTable = tempIdfTable.map(tf_idfInnerFunc)
    tempNormalizedCosTable = tempTf_idfTable.map(tempAggregateInnerTable) \
        .flatMap(lambda x: x) \
        .groupBy(lambda x: x[0]) \
        .mapValues(normalizeCos) \
        .map(lambda x: x[1]) \
        .flatMap(lambda x: x) \
        .groupBy(lambda x: x[1]) \
        .mapValues(makeDict)
        #.map(deleteNulls) \
        #.filter(lambda x: len(x[1]) != 0)
    tempVecCosTable = tempNormalizedCosTable.map(tempConvertToTable)
    
    return tempVecCosTable
```

```python
# Function that gets a string and return a list
# of ten nearest files

def TenNearestCos(string):
    tempVecCosTable = createNormCosTable(string)    
    
    vec = numpy.zeros(len(bookNames))
    TmpCosTable = tempVecCosTable.map(lambda x: tempMult(x, len(bookNames)))
    for i in range(len(bookNames)):
        vec[i] = TmpCosTable.map(lambda x: x[i]).sum()
    #print(vec)
        
    tempVecArr = vec.copy()
    fiveNearestIndxs = numpy.zeros(10)
    for i in range(len(fiveNearestIndxs)):
        fiveNearestIndxs[i] = numpy.argmax(tempVecArr).item()
        k = fiveNearestIndxs[i]
        tempVecArr[int(k.item())] = 0
        
    arrToReturn = list(fiveNearestIndxs)
    for i in range(len(arrToReturn)):
        arrToReturn[i] = int(arrToReturn[i])

    return arrToReturn
```

```python
# Function that devide the files into subsets anr returns the list of subsets

def kMeans():
    k1 = [0] * len(bookNames)
    k2 = [0] * len(bookNames)
    k3 = [0] * len(bookNames)
    randInxs = threeRandom(len(bookNames))
    
    for i in range(len(bookNames)):
        k1[i] = vecFull[randInxs[0]][i]#random.random()/5
        k2[i] = vecFull[randInxs[1]][i]#random.random()/5
        k3[i] = vecFull[randInxs[2]][i]#random.random()/5
    
    k1Dist = calcDistances(k1, vecFull)
    k2Dist = calcDistances(k2, vecFull)
    k3Dist = calcDistances(k3, vecFull)
    
    prevSubSets = [-1] * len(bookNames)
    curSubSets = upgradeSubSets(k1Dist, k2Dist, k3Dist)
    
    while (changedSubSets(curSubSets, prevSubSets)):
        k1Dist = calcDistances(k1, vecFull)
        k2Dist = calcDistances(k2, vecFull)
        k3Dist = calcDistances(k3, vecFull)
        
        prevSubSets = curSubSets.copy()
        curSubSets = upgradeSubSets(k1Dist, k2Dist, k3Dist)
        
        k1 = upgradeKPoint(curSubSets, 1, vecFull)
        k2 = upgradeKPoint(curSubSets, 2, vecFull)
        k3 = upgradeKPoint(curSubSets, 3, vecFull)
    
    return curSubSets
```

```python
# Function that gets a kPoint and a cos-distances matrix
# and return the list of distances from the kPoint to other files

def calcDistances(kPoint, vecArr):
    kDist = [0] * len(kPoint)
    
    for i in range(len(vecArr)):
        sqrSum = 0
        for j in range(len(vecArr[i])):
            sqrSum = sqrSum + (math.pow((kPoint[j] - vecArr[i][j]), 2))
        kDist[i] = math.sqrt(sqrSum)
    
    return kDist
```

```python
# Function that gets 3 kPoints and creates a new dividion to subsets of the files

def upgradeSubSets(k1Dist, k2Dist, k3Dist):
    newSubSets = [0] * len(bookNames)
    
    for i in range(len(newSubSets)):
        if (k1Dist[i] <= k2Dist[i]) and (k1Dist[i] <= k3Dist[i]):
            newSubSets[i] = 1
        elif (k2Dist[i] <= k1Dist[i]) and (k2Dist[i] <= k3Dist[i]):
            newSubSets[i] = 2
        elif (k3Dist[i] <= k1Dist[i]) and (k3Dist[i] <= k2Dist[i]):
            newSubSets[i] = 3
    
    return newSubSets
```

```python
# Function that upgrades the positon of the point

def upgradeKPoint(subSet, kPointNum, vecArr):
    newKPoint = [0] * len(bookNames)
    countSubSet = 0
            
    for i in range(len(vecArr)):
        for j in range(len(vecArr[i])):
            if (subSet[i] == kPointNum):
                newKPoint[j] = newKPoint[j] + vecArr[i][j]
                countSubSet = countSubSet + 1
                
    if (countSubSet != 0):
        for i in range(len(newKPoint)):
            newKPoint[i] = newKPoint[i] / countSubSet
    
    return newKPoint
```

```python
# Function that checks if list of subset was changed

def changedSubSets(curSubSets, prevSubSets):
    for i in range(len(curSubSets)):
        if (curSubSets[i] != prevSubSets[i]):
            return True
        
    return False
```

```python
# Function that returns three different numbers in a range

def threeRandom(rang):
    rand1 = random.randint(0, rang - 1)
    rand2 = random.randint(0, rang - 1)
    rand3 = random.randint(0, rang - 1)
    
    while(rand2 == rand1):
        rand2 = random.randint(0, rang)
        
    while(rand3 == rand1 or rand3 == rand2):
        rand3 = random.randint(0, rang)
        
    return (rand1, rand2, rand3)
```
